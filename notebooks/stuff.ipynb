```python
## Migrate the SQL DB to NOSQL 

import psycopg2
from util import web_scrape_content, web_scrape_tags

# Connect to your PostgreSQL database
conn = psycopg2.connect(
    host="localhost",
    database="postgres",
    user="postgres",
    password="password"
)

# Create a cursor
cur = conn.cursor()

# Execute a query
cur.execute("SELECT * FROM articles")

# Loop through all rows
for row in cur.fetchall():

    print(row)  # row is a tuple of column values

    document = {
        'date': row[2], 
        'title': row[3], 
        'link': row[4], 
        'source': row[5], 
        'ticker': row[1], 
        'tags': [],     
        'content': ''   
    }   
    

    collection.insert_one(document)
        

# Clean up
cur.close()
conn.close()
```
